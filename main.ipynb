```python
                                                                                            import sqlite3

# constants
from constants import CONNECTION_STRING
# functions
from category_functions import Category_Functions
from product_functions import Product_Functions
```

```python
# create connections
conn = sqlite3.connect(CONNECTION_STRING)
cur = conn.cursor()
```

```python
# create a category_functions instance which held all category functions
category_functions = Category_Functions(conn = conn, cur = cur)
```

```python
# create categories table
category_functions.create_categories_table()
```

```python
# get main categories
main_categories = category_functions.get_main_categories(save_db=True)
```

```python
# # get all categories
category_functions.get_all_categories(main_categories, save_db= True)
```

```python
# get all last level categories
last_level_categories = category_functions.get_all_last_level_categories()
```

```python
# create product functions
product_functions = Product_Functions(conn = conn, cur = cur)
```

```python
# create products table
product_functions.create_products_table()
```

```python
# get all products by last level categories
product_functions.get_all_products(last_level_categories, save_db = True)
```
